```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob
```

```python
# Load both files
df_parquet = pd.read_csv("../lightcurve_features_26_negative.parquet")
df_csv = pd.read_csv("../positive_100_elements.csv")
```

```python
df_parquet.shape
```

```python
df_csv.shape
```

```python
# concatenate negative and positive
combined = pd.concat([df_parquet, df_csv], ignore_index=True)

# Shuffle all rows
combined = combined.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
combined.shape
```

```python
# remove object name column
features_combined = combined.drop(columns=['object'])
```

```python
features_combined.shape
```

```python
# build correlation matrix of combined sample
corr_combined = features_combined.corr()
```

```python
corr_combined.shape
```

```python
# Plot combined correlation matrix

plt.figure(figsize=(12, 10))
sns.heatmap(corr_combined, cmap='coolwarm', center=0)
plt.title('FeatureFeature Correlation (baseline  matrix)')
plt.tight_layout()
plt.show()
```
