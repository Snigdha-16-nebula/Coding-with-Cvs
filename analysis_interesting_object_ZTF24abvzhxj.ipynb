```python
#import useful python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import io
```

```python
#download data from fink 
import requests
import pandas as pd
import io

# get data for ZTF24abvzhxj
r = requests.post(
    'https://api.ztf.fink-portal.org/api/v1/objects',
    json={
        'objectId': 'ZTF24abvzhxj',
        'output-format': 'json'
    }
)

# Format output in a DataFrame
pdf = pd.read_json(io.BytesIO(r.content))
```

```python
#dataframe for checking data
pdf
```

```python
#check columns
print(pdf.columns.tolist())
```

```python
#extracting columns of lightcurve data
df = pdf.copy()

df['jd'] = df['i:jd']
df['mag'] = df['i:magpsf']
df['err'] = df['i:sigmapsf']
df['fid'] = df['i:fid']
```

```python
#different filters
g = df[df['fid'] == 1]
r = df[df['fid'] == 2]
```

```python
# select r-band
#saving r band lightcurve data 
r_band = df[df['fid'] == 2].copy()

# keep only useful columns
r_band = r_band[['jd', 'mag', 'err']]

# save to CSV
r_band.to_csv('interesting_object_fink.csv', index=False)
```

```python
#showing the data
df_1=pd.read_csv('interesting_object_fink.csv')
```

```python
#FINK alert light curve

plt.figure()
#axis
plt.errorbar(g['jd'], g['mag'], yerr=g['err'], fmt='o', label='g')
plt.errorbar(r['jd'], r['mag'], yerr=r['err'], fmt='o', label='r')
#plot
plt.gca().invert_yaxis()
plt.xlabel("JD")
plt.ylabel("Magnitude")
plt.legend()
plt.title("Fink Alert Lightcurve")
plt.grid(True)
plt.show()
```

```python
# Saved data file
snad = pd.read_csv("interesting_object.csv")
fink = pd.read_csv("interesting_object_fink.csv")

#SNAD lightcurve
mjd = np.fromstring(snad['mjd_r'].iloc[0].strip('[]'), sep=' ')
mag = np.fromstring(snad['mag_r'].iloc[0].strip('[]'), sep=' ')
err = np.fromstring(snad['magerr_r'].iloc[0].strip('[]'), sep=' ')

#FINK r-band alert
fink = fink.copy()
fink['mjd'] = fink['jd']- 2400000.5 #converting mjd to jd
fink = fink.sort_values('mjd')

# PLOT two curves together
plt.figure()
#plot axis
plt.errorbar(mjd, mag, yerr=err, fmt='o', label='SNAD')
plt.errorbar(fink['mjd'], fink['mag'], yerr=fink['err'], fmt='s', label='FINK r')
#plot
plt.gca().invert_yaxis()
plt.xlabel("MJD")
plt.ylabel("Magnitude")
plt.legend()
plt.grid(True)
plt.title('ZTF24abvzhxj data release and alert')
# Save the figure
plt.savefig('ZTF24abvzhxj_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
